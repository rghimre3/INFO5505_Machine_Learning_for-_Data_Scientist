```python
df1= df.copy()
```

```python
bins = [0,5,10]
labels = [0, 1] # 'low'=0, 'high'=1
df1['quality_grade']= pd.cut(x=df1['quality'], bins=bins, labels=labels)

print(df1[['quality_grade','quality']].head(100))

df1 = df1.drop('quality', axis=1) 
```

```python

plt.figure(figsize = (20,22))
for i in range(1,12):
    plt.subplot(5,5,i)
    sns.distplot(df[df.columns[i]], fit=norm)
```

### **Logistic Regression Model**

```python
X = df1[[ 'alcohol', 'density', 'volatile_acidity', 'chlorides',
       'citric_acid', 'fixed_acidity', 'free_sulfur_dioxide',
       'total_sulfur_dioxide', 'sulphates', 'residual_sugar', 'pH']] 
y = df1.quality_grade
# split 80% data for training and 20% data for testing.
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=21)
```

```python
X.head(10)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
y.head(10)
```

```python
lr = LogisticRegression(random_state=21)
lr.fit(X_train, y_train)
```

### **Accuracy**

```python
train_accuracy = lr.score(X_train, y_train)
test_accuracy = lr.score(X_test, y_test)
print('Accuracy in Train Group   : {:.2f}'.format(train_accuracy), 
      'Accuracy in Test  Group   : {:.2f}'.format(test_accuracy), sep='\n')
```

```python
pred_test  = lr.predict(X_test)
pred_train = lr.predict(X_train)
```

```python
# confusion matrix
cm = confusion_matrix(y_test,pred_test)
cm
```

```python
quality_pred = LogisticRegression(random_state=40)
quality_pred.fit(X_train,y_train)
```

```python
confusion_matrix_train = confusion_matrix(y_train,pred_train)
confusion_matrix_test = confusion_matrix(y_test,pred_test)

print('Confusion Matrix Train Data', '**', confusion_matrix_train, sep='\n')
print('Confusion Matrix Test Data', '**', confusion_matrix_test, sep='\n')
```

```python
# Printing the accuracy of the model

print("The accuracy Score of Our Model     : ",  quality_pred.score(X_test, y_test))
```

```python

# Printing the precision socre  of the model
print("precision_score(): ",  precision_score(y_test, pred_test, average='micro'))
```
